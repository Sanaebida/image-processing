```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Convertit l'image en niveaux de gris Y
image = cv.imread("camman.jpg")
b,v,r = cv.split(image)         # rcupre 3 matrices d'octets spares
y = 0.299*r + 0.587*v + 0.114*b # opration matricielle
y = y.astype(np.uint8)          # convertit les rels en octets

# Calcule l'histogramme de l'image
histo = np.zeros(256, int)      # prpare un vecteur de 256 zros
for i in range(0,image.shape[0]):       # numre les lignes
    for j in range(0,image.shape[1]):   # numre les colonnes
        histo[y[i,j]] = histo[y[i,j]] + 1
print(histo)
plt.plot(histo)
plt.show()

# Calcule l'histogramme cumul hc
hc = np.zeros(256, int)         # prpare un vecteur de 256 zros
hc[0] = histo[0]
for i in range(1,256):
    hc[i] = histo[i] + hc[i-1]

# Normalise l'histogramme cumul
nbpixels = y.size
hc = hc / nbpixels * 255
print(hc)
plt.plot(hc)
plt.show()

# Utilise hc comme table de conversion des niveaux de gris
for i in range(0,y.shape[0]):       # numre les lignes
    for j in range(0,y.shape[1]):   # numre les colonnes
        y[i,j] = hc[y[i,j]]
cv.imshow("image de l'histogramme cumul", y)
cv.waitKey(0)
cv.destroyAllWindows()
```
